```python
class PersonalInfoManager:
    def __init__(self, name, age, city, hobbies):
        self.name = name
        self.age = age
        self.city = city
        self.hobbies = hobbies

    def display_info(self):
        print("\n📋 Personal Information")
        print("-" * 30)
        print(f"👤 Name     : {self.name}")
        print(f"🎂 Age      : {self.age}")
        print(f"🏙️ City     : {self.city}")
        print("🎯 Hobbies  :")
        for i, hobby in enumerate(self.hobbies, 1):
            print(f"   {i}. {hobby}")
        print("-" * 30)


# Example usage
if __name__ == "__main__":
    name = "Tripti"
    age = 21
    city = "Tilda"
    hobbies = ["Coding", "Reading", "Playing Chess"]

    manager = PersonalInfoManager(name, age, city, hobbies)
    manager.display_info()
```
